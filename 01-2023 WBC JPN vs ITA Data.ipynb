```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.mlb.com/world-baseball-classic/stats/'
page = requests.get(url)
soup=BeautifulSoup(page.text,'lxml')
```

```python
table=soup.find('table',{'bui-table is-desktop-sKqjv9Sb'})
table
```

```python
mydata_column=['TEAM','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','CS','AVG','OBP','SLG','OPS']
mydata_column
```

```python
tableValues=[]

for j in table.find_all('tr')[1:]:
     row_data = j.find_all('td')
     td_val=[y.text for y in row_data]
     tableValues.append(td_val)
     print(td_val)
```

```python
tableValues_2=[]

for page in range(2,13):  # 執行1~3頁
    # 連結網站
    response = requests.get('https://www.mlb.com/world-baseball-classic/stats/?page='+str(page))
    # HTML原始碼解析
    soup_2 = BeautifulSoup(response.text, "html.parser")
    # 取得所有class為post_title的<h3>
    table_2=soup_2.find('table',{'bui-table is-desktop-sKqjv9Sb'})

    print(f"====================第{str(page)}頁====================")
    
    for d in table_2.find_all('tr')[1:]:
        row_data_2 = d.find_all('td')
        td_val_2=[y.text for y in row_data_2]
        print(td_val_2)
        tableValues_2.append(td_val_2)
```

```python
final_list=tableValues+tableValues_2
final_list
```

```python
df=pd.DataFrame(final_list,columns=mydata_column)
```

```python
df_TPE=df.loc[df['TEAM'] == 'TPE']
df_TPE
```
